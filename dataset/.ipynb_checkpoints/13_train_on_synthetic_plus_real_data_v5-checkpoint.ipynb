```python
import cv2
from PIL import Image, ImageDraw, ImageFont
import glob
import random
import json
import os
import numpy as np
import matplotlib.pyplot as plt
import string
from datasets import Dataset

import albumentations
import torch

from transformers import AutoImageProcessor, AutoModelForObjectDetection
```

```python
def crop_drawing_to_content(drawing):
    """
    Crop the input drawing to its content, removing white margins.

    The function finds the non-white edges of the drawing and crops 
    the image to the boundaries of its content. It assumes that the 
    drawing is predominantly white (255 in RGB).

    Args:
    - drawing (numpy.ndarray): Input image in RGB format with values 
      ranging from 0 to 255. Dimensions should be (height, width, channels).

    Returns:
    - numpy.ndarray: Cropped image containing only the content of the 
      drawing without white margins.
    """
    x_min_crop = 0
    x_max_crop = drawing.shape[1]
    y_min_crop = 0
    y_max_crop = drawing.shape[0]
    
    for x_min in range(0, drawing.shape[1]):
        if drawing[:,x_min,:].mean() != 255:
            x_min_crop = x_min
            break
    
    for x_max in range(drawing.shape[1]-1, -1,-1):
        if drawing[:,x_max,:].mean() != 255:
            x_max_crop = x_max
            break
    
    for y_min in range(0, drawing.shape[0]):
        if drawing[y_min,:,:].mean() != 255:
            y_min_crop = y_min
            break
    
    for y_max in range(drawing.shape[0]-1, -1,-1):
        if drawing[y_max,:,:].mean() != 255:
            y_max_crop = y_max
            break

    return drawing[y_min_crop:y_max_crop,x_min_crop:x_max_crop]
```

```python
# Paths to the folders containing the different types of images.
drawings_folder = 'data/synthetic_data/drawings'
edges_folder = 'data/synthetic_data/edges'
nodes_folder = 'data/synthetic_data/nodes'

# Retrieve the list of file names in each folder.
drawings_files_names = os.listdir(drawings_folder)
edges_files_names = os.listdir(edges_folder)
nodes_files_names = os.listdir(nodes_folder)

# Lists to store processed images from each folder.
drawings_set = []
edges_set = []
nodes_set = []

# Process and store images from the drawings folder.
for file_name in drawings_files_names:
    # Read the image.
    img = cv2.imread(os.path.join(drawings_folder, file_name))
    
    # Convert the color from BGR to RGB format.
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Crop the image to remove white margins.
    img = crop_drawing_to_content(img)

    # Determine the resize dimensions such that the longer side is 100 pixels.
    if img.shape[0] > img.shape[1]:
        height = 100
        width = int( (height / img.shape[0]) * img.shape[1] )
    else:
        width = 100
        height = int( (width / img.shape[1]) * img.shape[0] )
    
    # Resize the image.
    img = cv2.resize(img, (width, height), interpolation = cv2.INTER_AREA)
    
    # Append the processed image to the drawings set.
    drawings_set.append(img)

# Process and store images from the edges folder.
for file_name in edges_files_names:
    img = cv2.imread(os.path.join(edges_folder, file_name))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = crop_drawing_to_content(img)
    edges_set.append(img)

# Process and store images from the nodes folder.
for file_name in nodes_files_names:
    img = cv2.imread(os.path.join(nodes_folder, file_name))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = crop_drawing_to_content(img)
    nodes_set.append(img)

# Append the last processed image to the drawings set (this seems to be a mistake).
drawings_set.append(img)
```

```python
def generate_random_bounding_box(image, drawing, min_scale=0.5, max_scale=2.0):
    """
    Generate a random bounding box for the drawing to be placed on the image.
    
    Args:
    - image (numpy.ndarray): The target image where the drawing will be placed.
    - drawing (numpy.ndarray): The drawing that needs to be placed on the target image.
    - min_scale (float, optional): Minimum scaling factor for resizing the drawing. Defaults to 0.5.
    - max_scale (float, optional): Maximum scaling factor for resizing the drawing. Defaults to 2.0.
    
    Returns:
    - list[int]: A list containing the top-left coordinates (x, y) of the bounding box and its dimensions (width, height).
    """
    
    width, height = image.shape[:2]
    drawing_width, drawing_height = drawing.shape[:2]

    scale_factor = random.uniform(min_scale, max_scale)

    scaled_width = int(drawing_width * scale_factor)
    scaled_height = int(drawing_height * scale_factor)

    x = random.randint(0, width - scaled_width)
    y = random.randint(0, height - scaled_height)

    return [x, y, scaled_width, scaled_height]
```

```python
def generate_synthentic_dataset(num_images, drawings_set, edges_set, nodes_set):
    """
    Generate a synthetic dataset with random placements of drawings, edges, and nodes on a blank canvas.
    
    Args:
    - num_images (int): Number of synthetic images to be generated.
    - drawings_set (list of numpy.ndarray): List of drawing images.
    - edges_set (list of numpy.ndarray): List of edge images.
    - nodes_set (list of numpy.ndarray): List of node images.
    
    Returns:
    - list[dict]: A list of dictionaries, where each dictionary represents the annotation 
                  for a synthetic image, containing the image itself, its dimensions, 
                  and the objects placed on it.
    """
    
    annotatios = []
    
    for i in range(num_images):

        canvas               = np.ones((1024, 1024, 3), dtype=np.uint8) * 255
        
        num_drawings         = random.randint(0, min(10, len(drawings_set)))
        num_edges            = random.randint(0, min(30, len(edges_set)))
        num_nodes            = random.randint(0, min(30, len(nodes_set)))
        
        drawings             = random.sample(drawings_set, num_drawings)
        edges                = random.sample(edges_set, num_edges)
        nodes                = random.sample(nodes_set, num_nodes)

        bbox     = []
        area     = []
        id       = []
        category = []
        
        for j in range(num_drawings):

            while(True):
            
                bb = generate_random_bounding_box(canvas, drawings[j])

                x_min = bb[0]
                x_max = bb[0]+bb[2]
                y_min = bb[1]
                y_max = bb[1]+bb[3]

                if canvas[y_min:y_max, x_min:x_max].mean() == 255:
                    
                    bbox.append(bb)
                    area.append(bb[2]*bb[3])
                    id.append((j+1))
                    category.append(0)
                    
                    canvas[y_min:y_max, x_min:x_max] = cv2.resize(drawings[j], (bb[2], bb[3]))
                    break

        for j in range(num_edges):

            while(True):
            
                bb = generate_random_bounding_box(canvas, edges[j])

                x_min = bb[0]
                x_max = bb[0]+bb[2]
                y_min = bb[1]
                y_max = bb[1]+bb[3]

                if canvas[y_min:y_max, x_min:x_max].mean() == 255:
                    
                    canvas[y_min:y_max, x_min:x_max] = cv2.resize(edges[j], (bb[2], bb[3]))
                    break

        
        for j in range(num_nodes):

            while(True):
            
                bb = generate_random_bounding_box(canvas, nodes[j])

                x_min = bb[0]
                x_max = bb[0]+bb[2]
                y_min = bb[1]
                y_max = bb[1]+bb[3]

                if canvas[y_min:y_max, x_min:x_max].mean() == 255:
                    
                    canvas[y_min:y_max, x_min:x_max] = cv2.resize(nodes[j], (bb[2], bb[3]))
                    break

        

        objects = {'id': id,
                   'area': area,
                   'bbox': bbox,
                   'category': category} 
                
        
        image_data = {
            "image_id": (i+1),
            "image":Image.fromarray(canvas),
            "width": 1024,
            "height": 1024,
            "objects":objects
        }

        annotatios.append(image_data)

    return annotatios
```

```python
train_test_split = 0.3

cutoff_index_drawing = int( len(drawings_set) * (1-train_test_split) )
cutoff_index_edges   = int( len(edges_set) * (1-train_test_split) )
cutoff_index_nodes   = int( len(nodes_set) * (1-train_test_split) )
    
drawings_set_train = drawings_set[:cutoff_index_drawing]
drawings_set_eval  = drawings_set[cutoff_index_drawing:]

edges_set_train = edges_set[:cutoff_index_edges]
edges_set_eval  = edges_set[cutoff_index_edges:]

nodes_set_train = nodes_set[:cutoff_index_nodes]
nodes_set_eval  = nodes_set[cutoff_index_nodes:]

annotations_train = generate_synthentic_dataset(1000, drawings_set_train, edges_set_train, nodes_set_train)
annotations_eval  = generate_synthentic_dataset(100, drawings_set_eval, edges_set_eval, nodes_set_eval)
```

```python
checkpoint = "hustvl/yolos-base"
image_processor = AutoImageProcessor.from_pretrained(checkpoint, size=1024, longest_edge=1024)

transform = albumentations.Compose(
    [
        albumentations.Resize(1024, 1024),
        #albumentations.RandomCrop(width=450, height=450),
        #albumentations.HorizontalFlip(p=1.0),
        #albumentations.RandomBrightnessContrast(p=1.0),
    ],
    bbox_params=albumentations.BboxParams(format="coco", label_fields=["category"]),
)
```

```python
def formatted_anns(image_id, category, area, bbox):
    """
    Format annotations for a given image in a structured manner.
    
    Args:
    - image_id (int): Identifier for the image.
    - category (list[int]): List of category identifiers for each annotation.
    - area (list[float]): List of area values for each annotation.
    - bbox (list[tuple]): List of bounding box coordinates for each annotation.
    
    Returns:
    - list[dict]: A list of dictionaries, where each dictionary contains 
                  the structured annotation for an object in the image.
    """
    
    annotations = []
    for i in range(len(category)):
        new_ann = {
            "image_id": image_id,
            "category_id": category[i],
            "isCrowd": 0,  # Assumes there's no crowd in any annotation
            "area": area[i],
            "bbox": list(bbox[i]),
        }
        annotations.append(new_ann)

    return annotations
```

```python
# transforming a batch
def transform_aug_ann(examples):
    image_ids = examples["image_id"]
    images, bboxes, area, categories = [], [], [], []
    for image, objects in zip(examples["image"], examples["objects"]):
        image = np.array(image.convert("RGB"))[:, :, ::-1]
        out = transform(image=image, bboxes=objects["bbox"], category=objects["category"])

        area.append(objects["area"])
        images.append(out["image"])
        bboxes.append(out["bboxes"])
        categories.append(out["category"])

    targets = [
        {"image_id": id_, "annotations": formatted_anns(id_, cat_, ar_, box_)}
        for id_, cat_, ar_, box_ in zip(image_ids, categories, area, bboxes)
    ]

    return image_processor(images=images, annotations=targets, return_tensors="pt")
```

```python
train = dataset_train.with_transform(transform_aug_ann)
eval = dataset_eval.with_transform(transform_aug_ann)
```

```python
def collate_fn(batch):
    """
    Custom collation function to process and combine individual samples into a batch.
    
    Args:
    - batch (list[dict]): A list of dictionaries. Each dictionary contains two keys:
        "pixel_values": Image tensor or array.
        "labels": Labels associated with the image.
    
    Returns:
    - dict: A dictionary containing the batched images and labels.
        "pixel_values": Batched and padded image tensors.
        "labels": List of labels for the batch.
    """
    pixel_values = [item["pixel_values"] for item in batch]
    #encoding = image_processor.pad(pixel_values, return_tensors="pt")
    #encoding = image_processor.preprocess(pixel_values, do_normalize=True, do_resize=False, return_tensors="pt")
    encoding = image_processor.pad(pixel_values, return_tensors="pt")
    labels = [item["labels"] for item in batch]
    batch = {}
    batch["pixel_values"] = encoding["pixel_values"]
    #batch["pixel_mask"] = encoding["pixel_mask"]
    batch["labels"] = labels
    return batch
```

```python
id2label={0:'drawing'}
label2id={'drawing':0}
```

```python
model = AutoModelForObjectDetection.from_pretrained(
    checkpoint,
    id2label=id2label,
    label2id=label2id,
    ignore_mismatched_sizes=True,
)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="training",
    per_device_train_batch_size=1,
    num_train_epochs=5,
    fp16=True,
    save_steps=200,
    logging_steps=100,
    learning_rate=1e-5,
    weight_decay=1e-4,
    save_total_limit=2,
    remove_unused_columns=False,
    push_to_hub=False,
    evaluation_strategy="steps"
)
```

```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=collate_fn,
    train_dataset=train,
    eval_dataset=eval,
    tokenizer=image_processor,
)

trainer.train()
```

```python
def get_datasets(annotations_file, images_folder):
    
    with open(annotations_file) as fp:
        annotations = json.loads(fp.read())
    
    elements = []
    
    for i, image_description in enumerate(annotations['images']):
    
        img = Image.open(os.path.join(images_folder, image_description['file_name']))
    
        objects = {'id':[], 
                   'area': [],
                   'bbox': [],
                   'category': []}
        
        for j, ann_description in enumerate(annotations['annotations']):
    
            if ann_description['image_id'] == image_description['id']:
    
                objects['id'].append(ann_description['id'])
                objects['area'].append(ann_description['area'])
                objects['bbox'].append(ann_description['bbox'])
                objects['category'].append(ann_description['category_id'])
        
        
        el = {'image_id': image_description['id'],
              'image': img,
              'width': image_description['width'],
              'height': image_description['height'],
              'objects': objects
        }
    
        elements.append(el)

    return Dataset.from_list(elements)
```

```python
dataset_train = get_datasets(train_annotations_file, train_images_folder)
dataset_test = get_datasets(test_annotations_file, test_images_folder)
```

```python
# transforming a batch
def transform_aug_ann(examples):
    image_ids = examples["image_id"]
    images, bboxes, area, categories = [], [], [], []
    for image, objects in zip(examples["image"], examples["objects"]):
        image = np.array(image.convert("RGB"))[:, :, ::-1]
        out = transform(image=image, bboxes=objects["bbox"], category=objects["category"])
       
        area.append(objects["area"])
        images.append(out["image"])
        bboxes.append(out["bboxes"])
        categories.append(out["category"])

    targets = [
        {"image_id": id_, "annotations": formatted_anns(id_, cat_, ar_, box_)}
        for id_, cat_, ar_, box_ in zip(image_ids, categories, area, bboxes)
    ]

    return image_processor(images=images, annotations=targets, return_tensors="pt")

def transform_not_aug_ann(examples):
    image_ids = examples["image_id"]
    images, bboxes, area, categories = [], [], [], []
    for image, objects in zip(examples["image"], examples["objects"]):
        image = np.array(image.convert("RGB"))[:, :, ::-1]
        out = not_transform(image=image, bboxes=objects["bbox"], category=objects["category"])
       
        area.append(objects["area"])
        images.append(out["image"])
        bboxes.append(out["bboxes"])
        categories.append(out["category"])

    targets = [
        {"image_id": id_, "annotations": formatted_anns(id_, cat_, ar_, box_)}
        for id_, cat_, ar_, box_ in zip(image_ids, categories, area, bboxes)
    ]

    return image_processor(images=images, annotations=targets, return_tensors="pt")
```

```python
def show_predictions(image, threshold=0.9):
    """
    Process the given image to detect objects and visualize the detections by 
    drawing bounding boxes around them.
    
    Args:
    - image (PIL.Image): The input image to process.
    - threshold (float, optional): Confidence threshold to consider a detection as valid. 
                                  Detections with scores below this will be ignored. 
                                  Default is 0.9.
    
    Returns:
    - PIL.Image: The processed image with visualized detections.
    """
    
    inputs = image_processor(images=image, return_tensors="pt")
    outputs = model(**inputs)
    
    target_sizes = torch.tensor([image.size[::-1]])
    results = image_processor.post_process_object_detection(outputs, target_sizes=target_sizes, threshold=threshold)[0]
    
    draw = ImageDraw.Draw(image)
    for score, label, box in zip(results["scores"], results["labels"], results["boxes"]):
        box = [round(i, 2) for i in box.tolist()]
        print(
                f"Detected {model.config.id2label[label.item()]} with confidence "
                f"{round(score.item(), 3)} at location {box}"
        )
        draw.rectangle(box, outline="red", width=1)
    
    return image
```
